```python
#Body Mass Index (BMI)
#Input weight and height
#Calculate and print BMI
```

```python
height = input("What is your height? ") #height in meters
weight = input("What is your weight? ") #weight in kilograms
name = input("What is your name? ")
BMI = int(weight)/int(height)
print("My name is ", name, "and my BMI is ", BMI)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# to install in VScode use pip install seaborn in terminal
# to install in VScode use pip install matplotlib in terminal
```

```python
#bar plot
sns.set_theme(style="ticks", color_codes=True)
titanic = sns.load_dataset("titanic")
sns.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic)
plt.show()
```

```python
#count plot
sns.set_theme(style="ticks", color_codes=True)
titanic = sns.load_dataset("titanic")
p1=sns.countplot(x="sex", data=titanic, hue="class")
p1.set_title("Plot for counting")
plt.show()
```

```python
#scatter plot
g = sns.FacetGrid(titanic, row="sex", hue="alone")
g=(g.map(plt.scatter, "age", "fare").add_legend())
plt.show
```
