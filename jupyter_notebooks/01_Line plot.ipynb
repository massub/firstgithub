### Adding titles

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
phool = sns.load_dataset("iris")
phool

# #draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Sepal length X Sepal width")
plt.show()
```

## Defining x and y limits

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
phool = sns.load_dataset("iris")
phool

# #draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Sepal length X Sepal width")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

## Set styles
-darkgrid
-whitegrid
-dark
-white
-ticks

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
phool = sns.load_dataset("iris")
phool

# to remove the previously set style
sns.set_style(style=None, rc=None)

# #draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Sepal length X Sepal width")
sns.set_style("whitegrid")
sns.set_style(style=None, rc=None) # to remove the previously set style
plt.show()
```

## Size of Figures

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
phool = sns.load_dataset("iris")
phool

# to remove the previously set style
sns.set_style(style=None, rc=None)

#change figure size
plt.figure(figsize=(8,10)) #you can play with figure size according to need

# #draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Sepal length X Sepal width")
sns.set_style("darkgrid")
plt.show()
```
